# Taschenrechner Version 4

In dieser Version des Taschenrechners soll der Taschenrechner um die Möglichkeit erweitert werden, in den Ausdrücken Klammern zu verwenden.

**Beispiel**

<pre>(55 * (3 + 7) - 50 ) / 50</pre>

Um diesen Ausdruck berechnen zu können, müssen zunächst die Klammern von den innersten zu den äußersten hin aufgelöst werden und der Inhalt der Klammern bevorzugt berechnet werden.

Wir gehen dazu wie folgt vor:

<pre>
(1) Die letzte öffnende Klammer im Ausdruck finden und die Position merken:
=> "(55 * (3 + 7) - 50 ) / 50"
          ^----------------------- die letzte öffnende Klammer ist an Position 6
(2) Die erste schließende Klammer nach der letzten öffnenden Klammer finden und die Position merken:
=> "(55 * (3 + 7) - 50 ) / 50"
                ^----------------- die zugehörige schließende Klammer ist an Position 12
(3) Den Ausdruck in drei Teile je vor der Klammer, in der Klammer und nach der Klammer aufteilen:
=> "(55 * "  |  "3 + 7"  |  " - 50 ) / 50"
(4) Den mittleren Teil mit calc() berechnen und in einen String umwandeln
=> "(55 * "  |  "10.0"  |  " - 50 ) / 50"
(5) Den Ausdruck wieder zusammensetzen und mit calc() berechnen
=> "(55 * 10.0 - 50 ) / 50"
</pre>

Im Aufruf von calc() in Schritt 5 wird dann das nächste Klammerpaar erkannt und entsprechend aufgeteilt.

**Aufgabe**

Erweitern Sie den Taschenrechner, so dass Klammern nach dem oben angegebenen Schema aufgelöst werden können. Beginnen Sie dabei mit der gegebenen Lösung aus dem letzten Taschenrechner Arbeitsblatt.

**Zusatz**

Der Taschenrechner kann in der hier vorgestellten Lösung nicht alle denkbaren Ausdrücke mit Klammern berechnen. Finden Sie durch Überlegung oder Experimente heraus, welche Ausdrücke nicht berechnet werden können und warum.

```python
def find_character(text, delimiter):
    for i in range(len(text)):
        if text[i] == delimiter:
            return i
    return -1

def calc(command):
    # hier nach den Operatoren suchen
    operators = ['+', '-', '*', '/']
    pos = -1
    for op in operators:
        pos = find_character(command, op)
        if pos != -1:
            operator = op
            break
    
    if pos == -1:
        return float(command)
    
    # hier den Ausdruck aufteilen und die Teilausdrücke berechnen
    lhs = calc(command[:pos])
    rhs = calc(command[pos+1:])

    # hier das Ergebnis berechnen und zurückgeben
    if operator == '+':
        return lhs + rhs
    if operator == '-':
        return lhs - rhs
    if operator == '*':
        return lhs * rhs
    if operator == '/':
        return lhs / rhs    

# Taschenrechner Hauptprogramm
command = input("Bitte Ausdruck eingeben: ")
ergebnis = calc(command)    
print("Ergebnis: " + str(ergebnis))
```

```python
def find_character(text, delimiter):
    for i in range(len(text)):
        if text[i] == delimiter:
            return i
    return -1

def calc(command):
    # hier Ausdrücke in Klammern auswerten:
    #    1: letzte öffnende Klammer finden (dazu command umdrehen und erste öffnende Klammer suchen)
    command_rev = command[-1::-1]
    start_pos = find_character(command_rev, '(')
    if start_pos != -1:
        start_pos = len(command) - start_pos
        #    2: erste schließende Klammer nach der letzten öffnenden Klammer finden
        ende_pos = find_character(command[start_pos:], ')') + start_pos + 1
        #    3: command in drei Teile zerlegen (vor Klammerpaar, in Klammer, nach Klammerpaar)
        lhs    = command[:(start_pos-1)]
        middle = command[start_pos:(ende_pos-1)]
        rhs    = command[ende_pos:]
        #    4: Teil in Klammer berechnen und wieder in String umwandeln
        middle = str(calc(middle))
        #    5: Ausdruck wieder zusammensetzen und berechnen
        command = lhs + middle + rhs
        return calc(command)
    
    # hier nach den Operatoren suchen
    operators = ['+', '-', '*', '/']
    pos = -1
    for op in operators:
        pos = find_character(command, op)
        if pos != -1:
            operator = op
            break
    
    if pos == -1:
        return float(command)
    
    # hier den Ausdruck aufteilen und die Teilausdrücke berechnen
    lhs = calc(command[:pos])
    rhs = calc(command[pos+1:])

    # hier das Ergebnis berechnen und zurückgeben
    if operator == '+':
        return lhs + rhs
    if operator == '-':
        return lhs - rhs
    if operator == '*':
        return lhs * rhs
    if operator == '/':
        return lhs / rhs    

# Taschenrechner Hauptprogramm
command = input("Bitte Ausdruck eingeben: ")
ergebnis = calc(command)    
print("Ergebnis: " + str(ergebnis))
```
