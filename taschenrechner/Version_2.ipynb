Wenn kein Operator im eingegebenen String gefunden wird, wird die Variable <code>operator</code> in Zeile 10 nicht initialisiert und in <code>pos</code> bleibt der Wert -1 stehen.

In diesem Fall soll das Programm versuchen, den eingegebenen Ausdruck direkt in eine Fließkommazahl umzuwandeln und in der Variable <code>ergebnis</code> zu speichern.

**Aufgabe**

Erweitern Sie das Programm oben entsprechend.

```python
command = input("Bitte Ausdruck eingeben: ")

# hier nach den Operatoren suchen
operators = ['+', '-', '*', '/']
pos = -1
for op in operators:
    for i in range(len(command)):
        if command[i] == op:
            pos = i
            operator = op
            break
    if pos != -1:
        break

# hier den Ausdruck aufteilen und in fließkommazahlen umwandeln
lhs = float(command[:pos])
rhs = float(command[pos+1:])

# hier das Ergebnis berechnen und in "ergebnis" speichern
if operator == '+':
    ergebnis = lhs + rhs
if operator == '-':
    ergebnis = lhs - rhs
if operator == '*':
    ergebnis = lhs * rhs
if operator == '/':
    ergebnis = lhs / rhs
    
print("Ergebnis: " + str(ergebnis))
```

```python
# hier Funktion find_character() implementieren
def find_character(text, delimiter):
    for i in range(len(text)):
        if text[i] == delimiter:
            return i
    return -1

# hier die Funktion calc() implementieren
def calc(command):
    # hier nach den Operatoren suchen
    operators = ['+', '-', '*', '/']
    pos = -1
    for op in operators:
        pos = find_character(command, op)
        if pos != -1:
            operator = op
            break
    
    if pos == -1:
        return float(command)
    
    # hier den Ausdruck aufteilen und in fließkommazahlen umwandeln
    lhs = float(command[:pos])
    rhs = float(command[pos+1:])

    # hier das Ergebnis berechnen und zurückgeben
    if operator == '+':
        return lhs + rhs
    if operator == '-':
        return lhs - rhs
    if operator == '*':
        return lhs * rhs
    if operator == '/':
        return lhs / rhs    

# Taschenrechner Hauptprogramm
command = input("Bitte Ausdruck eingeben: ")
ergebnis = calc(command)    
print("Ergebnis: " + str(ergebnis))
```
