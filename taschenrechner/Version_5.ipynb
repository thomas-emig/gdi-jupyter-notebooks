# Taschenrechner Version 5

In dieser Version wird der Taschenrechner nicht im Funktionsumfang erweitert. Stattdessen werden wir den Aufbau des Programmcodes verändern.

Die Operatoren sollen in dieser Aufgabe durch Instanzen von zuvor definierten Klassen ersetzt werden. Für jeden möglichen Operator muss also eine Klasse definiert werden. Die Klassen sollen die folgenden Elemente beinhalten:

- Eine Funktion <code>compute()</code>, die zwei Eingabeparameter annimmt und ein Ergebnis zurückliefert, dass der Verknüpfung der beiden Parameter mit dem entsprechenden Operator entspricht.
- Eine nicht gekapselte Eigenschaft <code>character</code>, in der das verwendete Zeichen des Operators gespeichert ist.

Für die Addition würde das wie folgt aussehen:

- Die Operation <code>compute()</code> gibt die Summe der beiden Eingabeparameter zurück.
- Die Eigenschaft <code>character</code> enthält das Zeichen <code>"+"</code>.

In der Funktion <code>calc()</code> sollen diese Operatoren anschließend entsprechend anstatt der bisherigen Operatoren verwendet werden.

**Aufgabe**

Erweitern Sie die unten stehende Angabe aus der Letzten Übung um die oben beschriebenen Operatoren.

```python
def find_character(text, delimiter):
    for i in range(len(text)):
        if text[i] == delimiter:
            return i
    return -1

def calc(command):
    # hier Ausdrücke in Klammern auswerten:
    #    1: letzte öffnende Klammer finden (dazu command umdrehen und erste öffnende Klammer suchen)
    command_rev = command[-1::-1]
    start_pos = find_character(command_rev, '(')
    if start_pos != -1:
        start_pos = len(command) - start_pos
        #    2: erste schließende Klammer nach der letzten öffnenden Klammer finden
        ende_pos = find_character(command[start_pos:], ')') + start_pos + 1
        #    3: command in drei Teile zerlegen (vor Klammerpaar, in Klammer, nach Klammerpaar)
        lhs    = command[:(start_pos-1)]
        middle = command[start_pos:(ende_pos-1)]
        rhs    = command[ende_pos:]
        #    4: Teil in Klammer berechnen und wieder in String umwandeln
        middle = str(calc(middle))
        #    5: Ausdruck wieder zusammensetzen und berechnen
        command = lhs + middle + rhs
        return calc(command)
    
    # hier nach den Operatoren suchen
    operators = ['+', '-', '*', '/']
    pos = -1
    for op in operators:
        pos = find_character(command, op)
        if pos != -1:
            operator = op
            break
    
    if pos == -1:
        return float(command)
    
    # hier den Ausdruck aufteilen und die Teilausdrücke berechnen
    lhs = calc(command[:pos])
    rhs = calc(command[pos+1:])

    # hier das Ergebnis berechnen und zurückgeben
    if operator == '+':
        return lhs + rhs
    if operator == '-':
        return lhs - rhs
    if operator == '*':
        return lhs * rhs
    if operator == '/':
        return lhs / rhs    

# Taschenrechner Hauptprogramm
command = input("Bitte Ausdruck eingeben: ")
ergebnis = calc(command)    
print("Ergebnis: " + str(ergebnis))
```

```python
class operator_add:
    def __init__(self):
        self.character = "+"
    def compute(self, a, b):
        return a + b

class operator_subtract:
    def __init__(self):
        self.character = "-"
    def compute(self, a, b):
        return a - b
    
class operator_multiply:
    def __init__(self):
        self.character = "*"
    def compute(self, a, b):
        return a * b
    
class operator_divide:
    def __init__(self):
        self.character = "/"
    def compute(self, a, b):
        return a / b

def find_character(text, delimiter):
    for i in range(len(text)):
        if text[i] == delimiter:
            return i
    return -1

def calc(command):
    # hier Ausdrücke in Klammern auswerten:
    #    1: letzte öffnende Klammer finden (dazu command umdrehen und erste öffnende Klammer suchen)
    command_rev = command[-1::-1]
    start_pos = find_character(command_rev, '(')
    if start_pos != -1:
        start_pos = len(command) - start_pos
        #    2: erste schließende Klammer nach der letzten öffnenden Klammer finden
        ende_pos = find_character(command[start_pos:], ')') + start_pos + 1
        #    3: command in drei Teile zerlegen (vor Klammerpaar, in Klammer, nach Klammerpaar)
        lhs    = command[:(start_pos-1)]
        middle = command[start_pos:(ende_pos-1)]
        rhs    = command[ende_pos:]
        #    4: Teil in Klammer berechnen und wieder in String umwandeln
        middle = str(calc(middle))
        #    5: Ausdruck wieder zusammensetzen und berechnen
        command = lhs + middle + rhs
        return calc(command)
    
    # hier nach den Operatoren suchen
    operators = [operator_add(), operator_subtract(), operator_multiply(), operator_divide()]
    pos = -1
    for op in operators:
        pos = find_character(command, op.character)
        if pos != -1:
            operator = op
            break
    
    if pos == -1:
        return float(command)
    
    # hier den Ausdruck aufteilen und die Teilausdrücke berechnen
    lhs = calc(command[:pos])
    rhs = calc(command[pos+1:])

    # hier das Ergebnis berechnen und zurückgeben
    return operator.compute(lhs, rhs)

# Taschenrechner Hauptprogramm
command = input("Bitte Ausdruck eingeben: ")
ergebnis = calc(command)    
print("Ergebnis: " + str(ergebnis))
```
