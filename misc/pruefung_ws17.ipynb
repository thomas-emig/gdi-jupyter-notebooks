```python
def SummeDerTeilbarenZahlen(f, divisor)
    summe = 0
    i == 0
    while i < len(f):
        if f(i) % divisor = 0:
            summe = summe + f[i]
        i++
    return summe
```

Auf der rechten Seite ist der Programmcode einer Python Funktion gegeben. Die Funktion gibt eine Zahlenfolge aus, deren Werte sich durch die Funktion Fib(i) berechnet.

Befüllen Sie die untenstehende Tabelle. Geben Sie in der korrekten Reihenfolge die Ausgabe an, die über die Aufrufe der Funktion print ausgegeben werden.

**Eingabe: b = 2**

*Hinweis*: Die Anzahl der vorgegebenen Zeilen in der Tabelle unten ist nur eine Hilfestellung. Die Anzahl der gefragten Ausgaben kann kleiner oder größer sein.

Reihenfolge | Ausgabe des jeweiligen Aufrufs von print
---|---
1|
2|
3|
4|
5|
6|
7|

```python
def Zahlenfolge():
    b = int(input("b"))
    out = str(fib(0))
    for i in range(1, b+1):
        out += ", " + str(fib(i))
    print(out)
    
def fib(i):
    print("in: " + str(i))
    if i<2:
        k = i
    else:
        k = fib(i-1) + fib(i-2)
    return k

Zahlenfolge()
```
